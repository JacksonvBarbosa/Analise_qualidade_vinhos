```python
import pandas as pd

# Carregar datasets
df_train = extract_csv_raw("winequality-red.csv", sep=";")
df_production = pd.read_excel("../data/production/xgb_production.xlsx")

# Remover coluna target dos dados de treino
df_train_features = df_train.drop(columns=["quality"], errors="ignore")

# Remover colunas de previsões dos dados de produção
df_prod_features = df_production.drop(columns=["previsoes", "probabilidades"], errors="ignore")

# Estatísticas descritivas
summary_train = df_train_features.describe().T
summary_production = df_prod_features.describe().T

# Comparação lado a lado
comparison = summary_train[["mean", "std"]].join(
    summary_production[["mean", "std"]],
    lsuffix="_train", rsuffix="_production"
)

# Diferença percentual das médias
comparison["mean_diff_%"] = (
    (comparison["mean_production"] - comparison["mean_train"]) / comparison["mean_train"]
) * 100

print(comparison)
```
